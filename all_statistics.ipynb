CALCULATING MEAN
1. Method Calculation'


```python
population = [150,160,170,165,155]
sample = [160,150,165]

pop_mean = sum(population)/len(sample)
sample_mean = sum(sample)/len(sample)
print("population sum mean is", pop_mean)
print("sample sum mean is", sample_mean)
```

2. Using NUMPY

```python
import numpy as np
data = [80,90,80,70]
airthemetic_mean = np.mean(data)
print("the mean is", airthemetic_mean)
```
